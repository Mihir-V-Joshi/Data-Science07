```python
def convert(x):
    if x is not np.nan:
        last = x[-1]
        if last == 'M':
            num = float(x.lstrip('€').rstrip('M')) * 1000000
        else:
            num = float(x.lstrip('€').rstrip('K')) * 1000
        return(num)
```

```python
fifa['Value'] = fifa['Value'].apply(lambda x: convert(x)).astype(float)
```

```python
fifa['Wage'] = fifa['Wage'].apply(lambda x: convert(x)).astype(float)
```

```python
#Checking Value and Wage Columns after the process
fifa[['Value', 'Wage']].head()
```

```python
fifa.info()
```

```python
fifa['Joined']=fifa['Joined'].str.split(', ').str[1]
fifa['Joined']=fifa['Joined'].astype(float)
#since it has nan it cannot be converted to int hence changing to float
```

```python
fifa['Contract Valid Until'] = pd.to_datetime(fifa['Contract Valid Until'])
```

```python
fifa['Height'] = fifa['Height'].str.replace("'", ".").astype(float)
fifa['Weight'] = fifa['Weight'].str.rstrip('lbs').astype(float)
fifa[['Height', 'Weight']].head()
```

```python
fifa['Release Clause']=fifa['Release Clause'].apply(lambda x: convert(x)).astype(float)
```

```python
fifa['Release Clause'].head()
```

```python
fifa.isnull().sum()
```

```python
# 4.Check for missing values and do a mean imputation where necessary.
```

```python
num_fifa=fifa.select_dtypes(include=np.number)
cat_fifa=fifa.select_dtypes(exclude=np.number)
```

```python
num_fifa.isnull().sum()
```

```python
# Mean imputation for all numerical values
for i in num_fifa.columns:
    fifa[i]=fifa[i].fillna(fifa[i].mean())
#     num_fifa[i]=num_fifa[i].fillna(fifa[i].mean())
```

```python
fifa.isnull().sum()
```

```python
cat_fifa.isnull().sum()
```

```python
fifa.isnull().sum()
```

```python
fifa.loc[(fifa.Value == 0) & (fifa.Club.isna() == False), 'Value'] = fifa['Value'].mean()
```

```python
fifa.loc[(fifa.Joined.isna() == True) & (fifa.Club.isna() == False), 'Joined'] = np.random.choice(fifa.Joined)
```

```python
fifa.loc[(fifa['Contract Valid Until'].isna() == True) & (fifa.Club.isna() == False), 'Contract Valid Until'] = np.random.choice(fifa['Contract Valid Until'])
```

```python
# Work Rate                      48
# Body Type                      48
# Position                       60
fifa[fifa['Position'].isnull()]
```

```python
fifa.groupby(by='Position')[['SlidingTackle','Acceleration']].mean()
ind=fifa[fifa['Position'].isnull()].index
q1=fifa['SlidingTackle'].quantile(0.25)
q3=fifa['SlidingTackle'].quantile(0.75)
for i in ind:
    if fifa.loc[i,'SlidingTackle'] < q1:
        fifa.loc[i,'Position']='ST'
    elif fifa.loc[i,'SlidingTackle'] > q1 and fifa.loc[i,'SlidingTackle'] < q3:
        fifa.loc[i,'Position']='CM'
    else:
        fifa.loc[i,'Position']='CB'
```

```python
ind=fifa[fifa['Work Rate'].isnull()].index
for i in ind:
    fifa.loc[i,'Work Rate'] ='Medium/ Medium'
```

```python
for i in ['Preferred Foot','Body Type']:
    fifa[i].fillna(fifa[i].mode()[0], inplace = True)
```

```python
fifa.loc[(fifa['Contract Valid Until'].isna() == True) & (fifa.Club.isna() == False), 'Contract Valid Until'] = np.random.choice(fifa['Contract Valid Until'])
```
